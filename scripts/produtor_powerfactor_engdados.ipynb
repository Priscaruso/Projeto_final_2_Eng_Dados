```python
#importa das bibliotecas necessárias
import boto3
import json
import time
import random
from datetime import datetime
```

```python
# produz os dados do sensor de 10 em 10 segundos

# seta valor inicial para variável id
id = 0

while 1:
  #cria o cliente para conexão com o serviço Kinesis da AWS
  cliente = boto3.client('kinesis', aws_access_key_id = 'AKIAQ4RUE4CP7USFIOPM', aws_secret_access_key = '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', region_name = 'us-west-2')

  #cria as variáveis para o registro
  id += 1
  ct = datetime.now()
  data_power = random.uniform(0.8, 1.0)


  #cria os dados em dicionários que vão ser transmitidos como JSON para o Kinesis Data Stream
  registro = {
          'id': f'{id}', 
          'data': f'{data_power}', 
          'type': 'powerfactor', 
          'timestamp': f'{ct}'
          }

# operação de gravar o registro no AWS Kinesis
  cliente.put_record(
      StreamName = 'Stream2',
      Data = json.dumps(registro),
      PartitionKey = '02'
  )

  time.sleep(10)
```
